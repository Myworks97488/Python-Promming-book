```python
#using a pickl module that reads and writes on the binary text file. let us start
import pickle

emp1 = {'Empno': 1201, 'Name': 'Anushree', 'age': 25, 'alary': 47000}
emp2 = {'Empno': 1211, 'Name': 'Zoya', 'age': 30, 'alary': 48000}
emp3 = {'Empno': 1251, 'Name': 'Simajeet', 'age': 27, 'alary': 49000}
emp4 = {'Empno': 1266, 'Name': 'alex', 'age': 29, 'alary': 50000}

empfile = open('Emp.dat', 'wb')
pickle.dump(emp1, empfile)
pickle.dump(emp2, empfile)
pickle.dump(emp3, empfile)
pickle.dump(emp4, empfile)

print("Successfully written four dictionaries")

empfile.close()

```
